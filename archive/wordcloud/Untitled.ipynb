```python
import pandas as pd
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
df = pd.read_csv("../../data/simpsons_normalizedtext.csv")
top50_df = pd.read_csv("../../data/top_50.csv")
```

```python
df = df.dropna()
characters = list(df['raw_character_text'].unique())

df.set_index("raw_character_text", inplace=True)
df['character_name'] = ""
```

```python
for character in characters:
    if '.' in character:
        new_name = character.replace('.','').lower()
        df.loc[[character],['character_name']] = new_name
    else:
        df.loc[[character],['character_name']] = character.lower()
```

```python
df
```

```python
top50 = {}

for row in top50_df.iterrows():
    character = row[1][0].lower()
    temp_df = df.loc[df['character_name']==character.replace('_',' ')]
    merged_text = ""
    
    for row_ in temp_df.iterrows():
        current_text = row_[1][1]
        merged_text = merged_text + " " + current_text

    top50[character] = merged_text
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

for character,text in top50.items():

    stopwords = set(STOPWORDS)
    stopwords.add("said")
    stopwords.add("im")
    stopwords.add("ill")

    if "simpson" in character.lower():
        mask = np.array(Image.open('masks/'+character.lower()+'.jpg'))
    
        image_colors = ImageColorGenerator(mask)
        wc = WordCloud(background_color="black", max_words=5000, mask=mask,stopwords=stopwords,
                    min_font_size=6, max_font_size=72 ,relative_scaling=0.9,color_func=image_colors)
    else:
        mask = np.array(Image.open('masks/sillhouette.jpg'))
        wc = WordCloud(background_color="black", max_words=5000,mask=mask,stopwords=stopwords,
                    relative_scaling=0.9,colormap="Wistia")

    wc.generate(text).to_file('../../wc/'+character.lower()+'_wc.jpg')
```
