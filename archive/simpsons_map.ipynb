```python
import pandas as pd 
```

```python
lines_df = pd.read_csv("../data/simpsons_script_lines.csv", low_memory=False)
episodes_df = pd.read_csv("../data/simpsons_episodes.csv")
loc_df = pd.read_csv("../data/simpsons_loc_outside_springfield.csv")
```

```python
df = pd.merge(loc_df,episodes_df,on="title",how="left")
```

```python
df.dropna(axis=0,subset=['id'])
```

```python
df
```

```python
df = df.drop(['original_air_date','production_code','imdb_rating','imdb_votes','image_url','video_url'],axis=1)
```

```python
locations=[]

for row in df.iterrows():
    country = row[1].country
    if country == "usa":
        country = "USA"
    else:
        country = country.title()
        
    if type(row[1].state)==float:
        if type(row[1].city)==float:
            location = ""
        else:
            location = row[1].city+', '
    elif type(row[1].city)==float:
        location = row[1].state+', '
    else:
        location = row[1].city+', '+row[1].state+', '
    
    locations.append(location.title()+country)

locations
```

```python
df['location']=pd.Series(locations)
df
```

```python
df.to_csv('simpsons_places.csv',index=False)
```

```python
loc_json = {}
for row in df.iterrows():
    lat = row[1].lat
    long = row[1].long
    location = row[1].location
    season = row[1].season
    episode = row[1].number_in_season
    title = row[1].title
    
    if location in loc_json:
        loc_json[location]['visits'].append({
            'title':title,
            'season':int(season),
            'episode':int(episode),
        })
    else:
        loc_json[location] = {
            'coordinates':[lat,long],
            'location':location,
            'visits':[{
                'title':title,
                'season':int(season),
                'episode':int(episode)
            }]
        }
```
